```python
from langchain.tools import tool
from langchain import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage, AIMessage, ToolMessage
from langchain_core.output_parsers import JsonOutputParser

from pydantic import BaseModel, Field
from typing import List
import requests
import json


# ---------------------------------------------------------
# TOOL DEFINITION
# ---------------------------------------------------------
@tool(description="Get nutrition information (calories, fat, etc.) for given recipe ingredients.")
def get_nutrition_info(recipe_ingredients: str) -> dict:
    """Get nutrition information for recipe ingredients."""
    api_url = 'https://api.calorieninjas.com/v1/nutrition?query='
    api_key =   Nutr_API_KEY# Replace with your actual key

    url = api_url + requests.utils.quote(recipe_ingredients)
    response = requests.get(url, headers={'X-Api-Key': api_key})

    if response.status_code != 200:
        return {
            "error": True,
            "status_code": response.status_code,
            "message": response.text
        }

    data = response.json()
    return {"items": data.get("items", [])}


# ---------------------------------------------------------
# STRUCTURED OUTPUT SCHEMA
# ---------------------------------------------------------
class NutritionInformation(BaseModel):
    calories: float | None = Field(None, description="Total calories")
    protein_g: float | None = Field(None, description="Protein in grams")
    fat_total_g: float | None = Field(None, description="Total fat in grams")
    carbohydrates_total_g: float | None = Field(None, description="Total carbohydrates in grams")


class RecipeInformation(BaseModel):
    name: str = Field(description="Recipe name")
    description: str = Field(description="Brief description of the recipe")
    ingredients: List[str] = Field(description="List of ingredients with quantities")
    nutrition_facts: NutritionInformation = Field(description="Nutritional information")
    instructions: List[str] = Field(description="Step-by-step cooking instructions")
    cooking_time_minutes: int = Field(description="Total cooking time in minutes")
    servings: int = Field(description="Number of servings")
    cuisine: str = Field(description="Cuisine type")
    dietary_restrictions: List[str] = Field(description="Dietary restrictions (e.g., vegetarian, gluten-free)")


# ---------------------------------------------------------
# SETUP MODEL
# ---------------------------------------------------------

# Model with tools for initial recipe generation and tool calling
model_with_tools = ChatGoogleGenerativeAI(
    model="gemini-2.5-pro",
    google_api_key=google_api_key,
    temperature=0.7
).bind_tools([get_nutrition_info])

# Model for final structured output
structured_model = ChatGoogleGenerativeAI(
    model="gemini-2.5-pro",
    google_api_key=google_api_key,
    temperature=0
)

# JSON output parser
parser = JsonOutputParser(pydantic_object=RecipeInformation)


# ---------------------------------------------------------
# PROMPT TEMPLATE
# ---------------------------------------------------------
prompt_template = PromptTemplate(
    input_variables=["cuisine"],
    template=(
        "Create a {cuisine} recipe with the following details:\n"
        "- Recipe name\n"
        "- Description\n"
        "- List of ingredients (with quantities)\n"
        "- Step-by-step instructions\n"
        "- Cooking time in minutes\n"
        "- Number of servings\n"
        "- Any dietary restrictions\n\n"
        "After creating the recipe, you MUST call the 'get_nutrition_info' tool "
        "with the ingredient list to get nutrition information."
    )
)


# ---------------------------------------------------------
# MAIN EXECUTION
# ---------------------------------------------------------
def generate_recipe_with_nutrition(cuisine: str):
    """Generate a recipe with nutrition information using tool calling."""
    
    # Step 1: Initial prompt
    initial_prompt = prompt_template.format(cuisine=cuisine)
    messages = [HumanMessage(content=initial_prompt)]
    
    # Step 2: Get AI response (should include tool calls)
    print("=== STEP 1: Initial AI Response ===")
    ai_msg = model_with_tools.invoke(messages)
    print(f"Tool calls: {len(ai_msg.tool_calls) if ai_msg.tool_calls else 0}")
    messages.append(ai_msg)
    
    # Step 3: Execute tools if present
    nutrition_data = None
    if ai_msg.tool_calls:
        print("\n=== STEP 2: Executing Tools ===")
        for tool_call in ai_msg.tool_calls:
            tool_name = tool_call["name"]
            args = tool_call["args"]
            
            if tool_name == "get_nutrition_info":
                print(f"Calling tool with args: {args}")
                # Call the tool - pass the string argument directly
                result = get_nutrition_info.invoke(args)
                print(f"Tool result: {result}")
                nutrition_data = result
                
                # Add tool result to messages
                messages.append(
                    ToolMessage(
                        content=str(result),
                        tool_call_id=tool_call["id"]
                    )
                )
    
    # Step 4: Generate final structured output using JSON formatting
    print("\n=== STEP 3: Generating Structured Output ===")
    
    # Get format instructions from parser
    format_instructions = parser.get_format_instructions()
    
    # Create a final prompt that includes all context
    final_prompt = f"""Based on the recipe and nutrition information above, 
provide the complete recipe details in JSON format.

{format_instructions}

Important: 
- Use the nutrition data from the tool call to fill in the nutrition_facts
- Ensure all fields are included
- Return ONLY valid JSON, no other text
"""
    messages.append(HumanMessage(content=final_prompt))
    
    # Get structured response
    response = structured_model.invoke(messages)
    
    # Parse the JSON response
    try:
        recipe_dict = parser.parse(response.content)
        final_response = RecipeInformation(**recipe_dict)
    except Exception as e:
        print(f"Parsing error: {e}")
        print(f"Raw response: {response.content}")
        # Try to extract JSON from response
        import re
        json_match = re.search(r'\{.*\}', response.content, re.DOTALL)
        if json_match:
            recipe_dict = json.loads(json_match.group())
            final_response = RecipeInformation(**recipe_dict)
        else:
            raise
    
    print("\n===== STRUCTURED OUTPUT =====")
    print(final_response)
    return final_response


# ---------------------------------------------------------
# EXAMPLE USAGE
# ---------------------------------------------------------
if __name__ == "__main__":
    recipe = generate_recipe_with_nutrition("Italian")
    
    print("\n===== FORMATTED RECIPE =====")
    print(f"Name: {recipe.name}")
    print(f"Cuisine: {recipe.cuisine}")
    print(f"Description: {recipe.description}")
    print(f"\nIngredients:")
    for ing in recipe.ingredients:
        print(f"  - {ing}")
    print(f"\nInstructions:")
    for i, step in enumerate(recipe.instructions, 1):
        print(f"  {i}. {step}")
    print(f"\nNutrition Facts:")
    print(f"  Calories: {recipe.nutrition_facts.calories}")
    print(f"  Protein: {recipe.nutrition_facts.protein_g}g")
    print(f"  Fat: {recipe.nutrition_facts.fat_total_g}g")
    print(f"  Carbs: {recipe.nutrition_facts.carbohydrates_total_g}g")
    print(f"\nCooking Time: {recipe.cooking_time_minutes} minutes")
    print(f"Servings: {recipe.servings}")
    print(f"Dietary Restrictions: {', '.join(recipe.dietary_restrictions) if recipe.dietary_restrictions else 'None'}")
```
